```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import os

# Define the model directory
model_dir = "/content/mbart_model"

# Create directory if it doesn't exist
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

# Load and save the model and tokenizer
model_name = "facebook/mbart-large-50-many-to-many-mmt"

# Download and save the tokenizer
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
tokenizer.save_pretrained(model_dir)

# Download and save the model
model = MBartForConditionalGeneration.from_pretrained(model_name)
model.save_pretrained(model_dir)

# Load the pre-trained model and tokenizer
tokenizer = MBart50TokenizerFast.from_pretrained(model_dir)
model = MBartForConditionalGeneration.from_pretrained(model_dir)

# Supported languages with their codes
languages = {
    'fr': 'fr_XX',  # French
    'es': 'es_XX',  # Spanish
    'de': 'de_DE',  # German
    'it': 'it_IT',  # Italian
    'pt': 'pt_XX',  # Portuguese
}

def translate(text, target_lang):
    """Translate text to the target language."""
    # Set source language as English (default)
    tokenizer.src_lang = 'en_XX'

    # Tokenize input text
    inputs = tokenizer(text, return_tensors="pt")

    # Generate translation
    target_lang_code = languages.get(target_lang)
    if target_lang_code is None:
        raise ValueError(f"Language '{target_lang}' is not supported.")

    translated_tokens = model.generate(
        **inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id[target_lang_code],
        num_beams=5,
        max_length=50,
        early_stopping=True
    )

    translation = tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

    return translation

def main():
    """Main function to handle user input and display translations."""
    print("Enter the text to translate into the following languages:")
    print("Supported Languages: French (fr), Spanish (es), German (de), Italian (it), Portuguese (pt)")

    text = input("Enter the text to translate: ").strip()
    if not text:
        print("Error: Text cannot be empty.")
        return

    # Translate to all supported languages
    translations = {}
    for lang_code in languages.keys():
        translations[lang_code] = translate(text, lang_code)

    # Display translations
    print("\nTranslations:")
    for lang_code, translation in translations.items():
        print(f"{lang_code}: {translation}")

if __name__ == "__main__":
    main()
```
